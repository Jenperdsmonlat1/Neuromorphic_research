```python
import numpy as np
import matplotlib.pyplot as plt
from LIF_model import LIF
from LIF_layer import LIF_layer
```

```python
dt = 1e-3
T = 1

neuron = LIF(dt=dt)
```

```python
synapse = [
    (10, [20e-3, 22e-3, 23e-3, 24e-3, 25e-3, 26e-3, 27e-3, 28e-3, 29e-3, 30e-3]),
    (50, [50e-3, 52e-3, 54e-3, 55e-3, 56e-3, 59e-3, 70e-3])]
```

```python
times = np.linspace(0, 1000, int(T / dt))

for t in range(int(T / dt)):
    neuron.steps(dt * t, synapse=synapse)
```

```python
print(int(T / dt))
print(times.shape)
print(len(neuron.V_trace))
```

```python
plt.figure()
plt.axhline(y=-0.056, color='r', linestyle='--', linewidth=2, label=f'Seuil = {-0.056} V')
plt.plot(times[0:100], neuron.V_trace[0:100])
plt.xlabel("Time (ms)")
plt.ylabel("V_membrane (v)")
plt.title("Potentiel membranaire au cours du temps.")
plt.legend()
plt.savefig("v_membrane.jpg")
plt.show()
```
