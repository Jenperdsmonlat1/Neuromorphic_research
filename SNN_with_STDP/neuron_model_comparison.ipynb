```python
import numpy as np
import matplotlib.pyplot as plt
from CUBA_LIF_layer import CUBA_LIF_Layer
from LIF_layer import LIF_Layer
from IF_layer import IF_Layer
from LIF_Sequential_layer import LIFSequential
from aquarel import load_theme
```

```python
cuba_lif_model = LIFSequential([
    CUBA_LIF_Layer(1, 1),
])

lif_model = LIFSequential([
    LIF_Layer(1, 1),
])

if_model = LIFSequential([
    IF_Layer(1, 1),
])
```

```python
time_step = 200

history_u_cuba_lif = []
history_u_lif = []
history_u_if = []

history_output_spikes_cuba_lif = []
history_output_spikes_lif = []
history_output_spikes_if = []
input_spikes = (np.random.rand(time_step, 1) > 0.85).astype(float)
```

```python
for t in range(time_step):
    _ = cuba_lif_model(input_spikes[t])
    history_u_cuba_lif.append(cuba_lif_model.layers[0].potentials.flatten().copy())
    history_output_spikes_cuba_lif.append(cuba_lif_model.layers[0].spikes.flatten().copy())
```

```python
history_u_cuba_lif = np.array(history_u_cuba_lif)
history_output_spikes_cuba_lif = np.array(history_output_spikes_cuba_lif)
```

```python
with load_theme("ambivalent").set_transforms(trim=True):

    plt.figure(figsize=(12, 6))

    plt.subplot(3, 1, 1)
    plt.stem(input_spikes)

    plt.subplot(3, 1, 2)
    plt.plot(history_u_cuba_lif, label='Neurone')

    plt.axhline(y=1.5, color='r', linestyle='--', label='Seuil (Threshold)')
    plt.title("Évolution du potentiel de membrane (Couche Cachée)")
    plt.xlabel("Pas de temps (ms)")
    plt.ylabel("Potentiel U")
    plt.legend(loc='upper right', fontsize='small', ncol=2)
    plt.grid(True, alpha=0.3)

    plt.subplot(3, 1, 3)
    plt.stem(history_output_spikes_cuba_lif, label='Neurone')

    plt.title("Spikes en sortie du neurone (Couche Cachée)")
    plt.xlabel("Pas de temps (ms)")
    plt.ylabel("Spikes émits")
    plt.legend(loc='upper right', fontsize='small', ncol=2)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
for t in range(time_step):
    _ = lif_model(input_spikes[t])
    history_u_lif.append(lif_model.layers[0].potentials.flatten().copy())
    history_output_spikes_lif.append(lif_model.layers[0].spikes.flatten().copy())
```
