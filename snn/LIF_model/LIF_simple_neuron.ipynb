```python
import numpy as np
import matplotlib.pyplot as plt
from SNNLayer import SNNLayer
```

```python
R = 10
dt = 0.20
T = 100
time = np.arange(0, T, dt)

tau = 20.0
E_L = 0.1
V_0 = 0.0

alpha = 1 - dt / tau
beta = dt * R / tau
```

```python
#spike_times = np.arange(10, T, 10)

N = 10
spike_times = np.sort(np.random.uniform(0, T, size=N))
input_spikes = np.zeros(len(time))

for t_spike in spike_times:
    print(t_spike)
    input_spikes[int(t_spike / dt)] = 1

print(spike_times)
print(input_spikes)
```

```python
v_trace = []
spike_out = []
V = 0.0

for n in range(len(time)):

    I = input_spikes[n]
    V = alpha * V + beta * I

    spike = 0
    if V >= E_L:
        spike = 1
        V = V_0
    
    v_trace.append(V)
    spike_out.append(spike)
```

```python
print(v_trace)
```

```python
plt.figure()
plt.plot(time, v_trace)
plt.show()
```

```python
plt.figure()
plt.plot(time, spike_out)
plt.show()
```
