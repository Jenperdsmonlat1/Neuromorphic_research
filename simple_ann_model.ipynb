```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
datas = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(datas.data, datas.target, test_size=0.2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

y_train_oh = tf.keras.utils.to_categorical(y_train, num_classes=3)
y_test_oh = tf.keras.utils.to_categorical(y_test, num_classes=3)
```

```python
inputs = tf.keras.layers.Input(shape=(4,))
x = tf.keras.layers.Dense(16, activation="relu")(inputs)
output = tf.keras.layers.Dense(3, activation="relu")(x)
model = tf.keras.models.Model(inputs, output)

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy'])
```

```python
model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10
)
```

```python
W_input_layer, b_input_layer = model.layers[1].get_weights()
W_output_layer, b_output_layer = model.layers[2].get_weights()
```

```python
np.save("W_input_layer.npy", W_input_layer)
np.save("W_output_layer.npy", W_output_layer)
```
